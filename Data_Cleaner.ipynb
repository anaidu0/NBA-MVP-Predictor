```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import os
import numpy as np
import urllib3
```

```python
# This method scrapes the MVP data from basketball-reference
def page_spider(): # defines URL and number of pages to crawl
    url = 'https://www.basketball-reference.com/awards/mvp.html' # enter URL to crawl
    source_code = requests.get(url) # gets url and stores it in a variable called source code
    soup = BeautifulSoup(source_code.content, 'html.parser') # uses all the website source code and assigns it to the soup object to utilize the BeautifulSoup module
    table2 = str(soup.find('table',{'class':'sortable stats_table'}))
    dfs = pd.read_html(table2)
    mvp_list = pd.DataFrame(dfs)
    mvp_list.to_csv("mvp.csv", index=False)
    #print (dfs)
```

```python
# Loads a downloaded data set. This dataset will be modified to be used in the model.
player_data = os.path.join('data','Seasons_Stats.csv')
```

```python
player_stats_df = pd.read_csv(player_data)
print(len(player_stats_df))
```

```python
sorted_stats_df = player_stats_df.sort_values('Player')
sorted_stats_df.head(20)
```

```python
#kept the relevant columns and exported it to a new file
keep_col = ['Year','Player','Age','G', 'MP', 'PER', 'TS%', 'WS', 'VORP', 'FG%', 'FT%', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PTS']
new_f = player_stats_df[keep_col]
new_f.to_csv("edited_stats.csv", index=False)
```

```python
#Converted all the total statistics to per game stats

ppg = new_f['PTS']/new_f['G']
ppg_series = pd.Series(ppg)
new_f['PPG'] = ppg_series.values

rpg = new_f['TRB']/new_f['G']
rpg_series = pd.Series(rpg)
new_f['RPG'] = rpg_series.values

apg = new_f['AST']/new_f['G']
apg_series = pd.Series(apg)
new_f['APG'] = apg_series.values

spg = new_f['STL']/new_f['G']
spg_series = pd.Series(spg)
new_f['SPG'] = spg_series.values

bpg = new_f['BLK']/new_f['G']
bpg_series = pd.Series(bpg)
new_f['BPG'] = bpg_series.values

new_f.to_csv("edited_stats.csv", index=False)
```

```python
df3 = new_f[new_f['Year'] >= 1985]  
```
